## **Data Exploration**

```python
#Four diff datasets: Glasses, smartphone, smartwatch and report
# Data can be found at https://www.kaggle.com/sasanj/human-activity-smart-devices
glasses_df = pd.read_csv(os.getcwd()+os.sep+'data'+os.sep+'glasses.csv')
watch_df = pd.read_csv(os.getcwd()+os.sep+'data'+os.sep+'smartwatch.csv')
phone_df = pd.read_csv(os.getcwd()+os.sep+'data'+os.sep+'smartphone.csv')
report_df = pd.read_csv(os.getcwd()+os.sep+'data'+os.sep+'report.csv')
pd.set_option('display.max_columns', None)
```

```python
glasses_df.sort_values(by=['DATE'],ascending=False, inplace= True)
glasses_df = glasses_df[["DATE","ACC_X","ACC_Y","ACC_Z","GYRO_X","GYRO_Y","GYRO_Z","EOG_L","EOG_R","EOG_H","EOG_V"]]
glasses_df.head(10)
```

```python
watch_df.sort_values(by=['timestamp'],ascending=False, inplace= True)
watch_df = watch_df[["timestamp","source","values"]]
watch_df.head(10)
```

```python
phone_df.sort_values(by=['timestamp'],ascending=False, inplace= True)
phone_df = phone_df[["timestamp","source","values"]]
phone_df.head(10)
```

```python
report_df.info()
report_df = report_df[["activity_type","duration","from","to"]]
```

### **Data preprocessing**
- Merging smartwatch and phone data
- Casting timestamp as datetime 

```python
#Merging phone and watch data
frames = [watch_df, phone_df]
device_df = pd.concat(frames)
device_df.head(10)
```

```python
device_df.info()
```

```python
glasses_df['timestamp'] = glasses_df['DATE']
glasses_df = glasses_df[["timestamp","ACC_X","ACC_Y","ACC_Z","GYRO_X","GYRO_Y","GYRO_Z","EOG_L","EOG_R","EOG_H","EOG_V"]]
```

```python
device_df['timestamp'] = pd.to_datetime(device_df['timestamp'])
```

```python
report_df['from'] = pd.to_datetime(report_df['from'])
report_df['to'] = pd.to_datetime(report_df['to'])
report_df.info()
report_df.head(10)
device_df.info()
device_df.head(10)
```

### **Data labelling**
Using report data to label activity type with sensor data. Sampling to make data labeling complete in resonable time.

```python
device_df['activity_type'] = ''
sample_df = device_df.sample(frac=0.1,replace=True, random_state=1)
for report_index, report_row in report_df.iterrows():
    for device_index, device_row in sample_df.iterrows():
        if report_row['from'] <= device_row['timestamp'] <= report_row['to']:
            sample_df.loc[device_index, "activity_type"] = report_row['activity_type']
```

```python
sample_df.head(10)
```

```python
sample_df.info()
```

```python
filtered_df = sample_df.copy()
filtered_df.drop(filtered_df[filtered_df['activity_type'] == ''].index, inplace=True)
```

### **Classification**

```python
filtered_df.head(10)
```

```python
X = filtered_df[['source','values']]
y = filtered_df['activity_type']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['source'] = le.fit_transform(X['source'])
X['values'] = le.fit_transform(X['values'])
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred  =  classifier.predict(X_test)
from sklearn.metrics import accuracy_score
ac = accuracy_score(y_test,y_pred)
```

```python
print("Accurracy:",ac)
```
